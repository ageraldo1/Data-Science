# GroupBy

```python
import pandas as pd
```

```python
fortune = pd.read_csv('../resources/fortune1000.csv', index_col=['Rank'])

fortune.head(3)
```

```python
# Always GroupBy using a column where exists duplicate data
# Let's use Sector column

fortune.groupby(by=['Sector'])
```

```python
sectors = fortune.groupby(by=['Sector'])
```

### 3. Methods on the GroupBy Object and DataFrame Columns

```python
# max function
# will return the max value for the most column to the left
# in this example, Company is the most column to the left
# Pandas will apply and sort descending order by Company Name in order to return the max of a String column
# The max function is applied to each group member(Sector field)

sectors.max().head(5)
```

```python
# min function

sectors.min().head(5)
```

```python
# sum function
# only applied to numerical columns

sectors.sum()
```

```python
# mean function

sectors.mean().head(3)
```

```python
# using a single series 
# sum of revenue per sector

sectors['Revenue'].sum()
```

```python
sectors['Profits'].min()
```

### 4. Grouping by Multiple Columns

```python
fortune.head(3)
```

```python
sectors = fortune.groupby(['Sector', 'Industry'])

sectors.size()
```

```python
# sum applied only for the numerical fields
# the dataframe returned is a multi-index dataframe

sectors.sum()
```

```python
sectors['Profits'].min()
```

### 5. The .agg() Method

```python
fortune.head(3)
```

```python
# instead of doing unique operations on the group (such as min, max, mean), using aggs you can apply different operation per column
# applying three different aggregations for different columns

sectors.agg({
    'Revenue' : 'sum',
    'Profits' : 'min',
    'Employees' : 'count'
}).head(3)
```

```python
# applying a list of operations 

sectors.agg(['size', 'min', 'mean', 'max']).head(10)
```

```python
# combining all operations together

sectors.agg({
    'Revenue' : ['min','mean', 'max'],
    'Profits' : 'min',
    'Employees' : 'size'
}).head(3)
```

### 6. Iterating through Groups

```python
# getting the max revenue per company and store the results into a new dataframe

# creating a empty Dataframe
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
# looping
for sector,data in sectors:
    highest_revenue = data.nlargest(n=1, columns='Revenue')
    df = df.append(highest_revenue)
    
df.head(5)
```

```python
# using the location column

df = pd.DataFrame(columns = fortune.columns)
cities = fortune.groupby(by='Location')

for city,data in cities:
    highest_profits = data.nlargest(n=1, columns='Profits')
    df = df.append(highest_profits)
    
df.head(5)
```
