### 8. Fill in Null Values with .fillna() Method

### 10. Sort a DataFrame with the .sort_values()

```python
nba = pd.read_csv('../resources/nba.csv')

nba.head(3)
```

```python
# sort Series is easy
#
# but for DataFrames...

nba.sort_values(by='Salary', ascending=False).head(3)
```

```python
# NaN values are put in the end

nba.sort_values(by='Salary', ascending=False, na_position='last').tail(3)
```

### 11. Sort a DataFrame with .sort_values using multiple columns

```python
nba.sort_values(by=['Team','Name'], ascending=True).head(3)
```

```python
# you can also change the ascending order per each sortBy field

nba.sort_values(by=['Team','Name'], ascending=[True, False]).head(5)
```

### 12. Sort DataFrame with .sort_index() Method

```python
# let's first mass our list

nba.sort_values(['Number', 'Salary', 'Name'], inplace=True)
nba.head(5)
```

```python
# let's put it back to the original order sorting by index!

nba.sort_index(inplace=True)
nba.head(5)   
```

### 13. Rank Values with the .rank() Method

```python
# you need first to remove NaN values in order for this method works

nba.dropna(how='any',inplace=True)
nba['Salary'] = nba['Salary'].astype(int)
```

```python
# Rank method only works for a single Series!

nba['Salary Rank'] = nba['Salary'].rank(ascending = False).astype(int)
nba.head(10)
```

```python
nba.sort_values(by='Salary Rank', ascending=True).head(10)
```

```python
nba.sort_values(by='Salary', ascending=False).head(10)
```

# Filtering

```python
df = pd.read_csv('../resources/employees.csv')

df.head(3)
```

```python
# optimization - first step for any analysis

df.info()
```

```python
# NaN values
# Gender can be changed from object to enum
# Team is a boolean value
# Start Date, Last Login time are not timestamp fields
# pay attention for the memory utilization by the object
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype(bool)
df['Gender'] = df['Gender'].astype('category')

df.info()
```

```python
df.head(10)
```

```python
# you can also change the dates files during the load

df = pd.read_csv('../resources/employees.csv',parse_dates = ['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype(bool)
df['Gender'] = df['Gender'].astype('category')

df.head(10)
```

```python
df.info()
```

### 14. Filter a DataFrame Based on A Condition

```python
# option 1 - using boolean Series

df[df['Gender'] == 'Male'].head(10)
```

```python
# negation

df[df['Team'] != 'Marketing'].head(3)
```

```python
# greater and less than

df[df['Salary'] >= 10000].head(3)
```

```python
df[df['Bonus %'] < 2.5].head(3)
```

```python
# dates
df[df['Start Date'] <= '1985-01-01'].head(3)
```

### 15. Filter with More than One Condition

```python
# store each condition in a variable
# all we need is a boolean Series to each condition. If the field is boolean, just plug the field direct

cond1 = df['Gender'] == 'Male'
cond2 = df['Team'] == 'Marketing'

df[cond1 & cond2].head(3)   # AND
```

```python
df[cond1 | cond2].head(3)   # OR
```

```python
# example 3

cond1 = df['First Name'] == 'Robert'
cond2 = df['Team'] == 'Client Services'
cond3 = df['Start Date'] > '2016-06-01'

df[(cond1 & cond2) | cond3]
```

### 16. The isin() method()

```python
# instead of using boolean series
# works like where column in (1,2,3) 
# works with Series as well

cond = df['Team'].isin(['Legal', 'Sales', 'Product'])

df[cond].head(10)
```

### 17. The isnull() and .notnull() Methods

### 18. The .between() Method

### 19. The .duplicated() Method

```python
df.sort_values(by='First Name', inplace=True)

df.head(5)
```

```python
# this list will return only the duplicated records
# Applied to Series!

df[df['First Name'].duplicated()].head(5)
```

```python
# negate the duplicate to get only the unique values (~ negate the boolean true)

df[~df['First Name'].duplicated()].head(5)
```

### 20 .drop_duplicates() Method

```python
# Applied to Dataframes! Passing a list as the subset parameter
# You need to specific the columns otherwise only when the entire row is duplicated

df.drop_duplicates(subset=['First Name'], keep='last').head(10)
```

### 23. Retrieve Rows by Index Label with .loc[]

```python
# importing again the CSV file but setting Film as the index column this time
# always when possible, try to keep the datasource sorted. 
# Very useful for String indexes

bond = pd.read_csv('../resources/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)

bond.head(3)
```

```python
# loc uses []
# use single records returns a Series

bond.loc['Goldfinger']
```

```python
# multiple records returns a Dataframe
bond.loc['Casino Royale']
```

```python
# using ranges
# end values are included

bond.loc['Diamonds Are Forever' : 'Moonraker']
```

```python
# passing a list

bond.loc[['Diamonds Are Forever','Moonraker', 'Movie that does not exists']]
```

```python
# checking if a movie exists

"Gold Bond" in bond
```
